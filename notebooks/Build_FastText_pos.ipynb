### Build Pos embedding from average of last two syllables

```python
import pandas as pd
import os
import fasttext
path = '../data/sanskrit_treebank/'
model = fasttext.load_model("../../fastText/result/cc.sanskrit.100.bin")
```

```python
pos = dict()
def clean_ma(ma):
    ma = re.sub("([\(\[]).*?([\)\]])", "\g<1>\g<2>", ma).replace('[] ','').strip(' []').replace(' ac','').replace(' ps','').replace('sgpl','sg').replace('sgdu','sg')
    ma = ma.replace('i.','inst.').replace('.','').replace(' ','')
    return ma
```

```python
import re
sub_folders = os.listdir(path)
for sub_folder in sub_folders:
    folder = os.path.join(path, sub_folder)
    files = sorted(os.listdir(folder))
    for file in files:
        full_path = os.path.join(folder, file)
        data = pd.read_csv(full_path, sep=',')
        for i in range(len(data)):
            ma = clean_ma(data.iloc[i,3])
            if ma not in pos.keys():
                pos[ma] = []
                pos[ma].append(data.iloc[i,1])
            else:
                pos[ma].append(data.iloc[i,1])
```

```python
count = 0
import numpy as np

import syllable_encoding
from indic_transliteration import sanscript
from indic_transliteration.sanscript import SchemeMap, SCHEMES, transliterate
f = open("pos_embedding_FT.100", "w")   
f.write(str(len(pos.keys()))+' '+str(100)+'\n')
for key in pos.keys():
    sums = np.zeros(100)
    for w in pos[key]:
        text = transliterate(w, sanscript.SLP1, sanscript.DEVANAGARI)
        w = transliterate(''.join(syllable_encoding.syllabify(text)[-2:]), sanscript.DEVANAGARI, sanscript.SLP1)
#         print(w)
        sums = sums + model.get_word_vector(w)
    sums = sums/len(pos[key])
    key=key+'@'
    print(key)
    f.write(key) 
    for v in sums:
        f.write(' '+str(v))
    f.write('\n')
f.close()
print('Done!!!')
```
