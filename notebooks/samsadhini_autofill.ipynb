```python
from mechanize import Browser
import shutil
import os
import pandas as pd
```

```python
# Get data from files:
path = '../Documents/Parsing/sanskrit_tree_bank/Preprocessed_amrith_data/Deliverables/1k_conll_sent_test/'
files = os.listdir(path)
count = 0
for file in files:
    data = pd.read_csv(path+file,sep=',')
    sentence= ' '.join(data['word'])
    name = file.replace('.csv','')
    ###################################################################3
    br = Browser()
    br.open("http://172.26.175.142/scl/SHMT/index.html")
    br.select_form(name="test")
    br['encoding'] = ['SLP']
    br['text']= sentence
    br.submit()
    
    # Give permissions to all the files
    !echo hk | sudo -S chmod a+rwx ./SKT_TEMP/*
    # Move newly generated file to store folder after renaming it
    all_subdirs = [d for d in os.listdir('./SKT_TEMP')]
    for dirs in all_subdirs:
        if 'tmp_in' in dirs:
            os.rename('./SKT_TEMP/'+dirs,'./SKT_TEMP/'+name)
            break
    shutil.move('./SKT_TEMP/'+name,'./SKT_TEMP/store/')  
    files = os.listdir('./SKT_TEMP/')
    for f in files:
        if (f.startswith("in")):
            shutil.move('./SKT_TEMP/'+f, './SKT_TEMP/'+name)
        else:
            pass
    ########################################################
    count=count+1
    print(count)
    if count>5:
        break
data
```
